 Data Visualization

Dataset: Iris Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
iris = pd.read_csv('Iris.csv')
feature_columns_from_csv = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
sklearn_feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
df = iris[feature_columns_from_csv]
df.columns = sklearn_feature_names
df.head()
```

```python
df['species'] = iris.target
df['species'] = df['species'].map(dict(enumerate(iris.target_names)))
df_grouped = df.groupby('species').mean()
```

```python
plt.figure()
df_grouped.plot(kind='bar')
plt.title("Average Feature Values per Species")
plt.xlabel("Species")
plt.ylabel("Value")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.hist(df['sepal length (cm)'])
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.tight_layout()
plt.show()
```
